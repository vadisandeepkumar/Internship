```python
#python program to display list of respected former presidents of India(

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
res=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
res
```

```python
soup=BeautifulSoup(res.content)
soup
```

```python
#Presidents
president=soup.find_all('div', class_='presidentListing')
president
```

```python
President_name=[]
for i in soup.find_all('div',class_="presidentListing"):
    President_name.append(i.find("h3").text.split("(")[0])

President_name
```

```python
years_of_service=[]
service=soup.find_all('div',class_="presidentListing")
for i in service: 
    years_of_service.append(i.find("p").text.split("(")[0])

years_of_service
```

```python
data = pd.DataFrame({"President Name":President_name,"Term of Office":years_of_service})
data
```
