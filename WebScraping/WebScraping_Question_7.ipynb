```python
#program to scrape mentioned news details from https://www.cnbc.com/world/?region=world
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
res= requests.get('https://www.cnbc.com/world/?region=world')
soup = BeautifulSoup(res.content)
```

```python
# 1. Headline
Headline=[]
h_line=soup.find_all('a',class_="LatestNews-headline")
for i in h_line:
    Headline.append(i.text)

Headline
```

```python
#2 Time
Time=[]
t_stamp=soup.find_all("time",class_="LatestNews-timestamp")
for j in t_stamp:
    Time.append(j.text)
    
Time
```

```python
# 3 NEWS LINK
News_Link=[]
N_Link=soup.find_all("a",class_="LatestNews-headline")
for k in N_Link:
    News_Link.append(k['href'])

News_Link
```

```python
data=pd.DataFrame({'HeadLine':Headline,'Time':Time,'News Link':News_Link})
data
```
