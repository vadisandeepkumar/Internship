```python
#program to scrape the details of most downloaded articles from AI in last 90 days
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
res= requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup = BeautifulSoup(res.content)
```

```python
paper=[]
author=[]
pubished_date=[]
url=[]
```

```python
p=soup('h2',class_="sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR")
for i in p:
    paper.append(i.text)
paper
```

```python
a=soup('span',class_="sc-1w3fpd7-0 pgLAT")
for j in a:
    author.append(j.text)
author
```

```python
p=soup('span',class_="sc-1thf9ly-2 bKddwo")
for i in p:

    pubished_date.append(i.text)
pubished_date
```

```python
u=soup('a',class_="sc-5smygv-0 nrDZj")
for l in u:
    url.append(l.get("href"))
    
url
```

```python
download= pd.DataFrame({"Paper Title":paper,"Authors":author,"Published Date":pubished_date,"Paper url":url})
download 
```
